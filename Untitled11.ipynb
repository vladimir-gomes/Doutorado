````python
import numpy as np
import torch
import torch.nn as nn
````

````python
# Step 1: Define the data matrix (Latitude, Longitude, Temperature)
data_matrix = np.array([
    [40.7, -74.0, 12.8],   # New York
    [34.0, -118.2, 18.6],  # Los Angeles
    [41.8, -87.6, 10.6],   # Chicago
    [25.8, -80.2, 24.9],   # Miami
    [47.6, -122.3, 11.3],  # Seattle
    [39.7, -104.9, 10.4],  # Denver
    [29.8, -95.4, 20.6],   # Houston
    [42.3, -71.0, 10.8],   # Boston
    [37.8, -122.4, 14.6],  # San Francisco
    [33.4, -112.0, 23.9]   # Phoenix
], dtype=np.float32)
````

````python
print (data_matrix)
````

````python
# Step 2: Extract features (X) and target (y)
X = data_matrix[:, :2]  # First two columns: Latitude, Longitude
y = data_matrix[:, 2]   # Last column: Temperature
````

````python
# Step 3: Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)
````

````python
# Step 4: Define the neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(2, 10)  # 2 inputs (Latitude, Longitude), 10 hidden units
        self.fc2 = nn.Linear(10, 1)  # 10 hidden units to 1 output (Temperature)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
````

````python
# Initialize model, loss function, and optimizer
model = SimpleNN()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
````

````python
# Step 5: Train the model
epochs = 1000
for epoch in range(epochs):
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')
````

````python
# Step 6: Make a prediction
new_location = torch.tensor([[45.0, -90.0]], dtype=torch.float32)
predicted_temp = model(new_location)
print(f'Predicted temperature for location (45.0, -90.0): {predicted_temp.item():.2f}°C')
````

````python
Claro\! Sem problemas. Aqui está o código integral, combinando todas as etapas em um único bloco, desde o carregamento dos dados do seu Google Drive até a análise final com todos os gráficos.

Copie e cole este bloco de código inteiro no seu ambiente (como Google Colab) e lembre-se de **alterar o caminho do arquivo** na seção "Etapa 1" para apontar para a sua tabela de dados.

### Código Integral: PINN para Desmistura Espectral a partir de Dados do Drive

```python
# -------------------------------------------------------------------
#                   *** BIBLIOTECAS NECESSÁRIAS ***
# -------------------------------------------------------------------
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.metrics import mean_squared_error

# ===================================================================
# ETAPA 1: MONTAR O GOOGLE DRIVE E CARREGAR SUA TABELA DE DADOS
# ===================================================================
print("--- Iniciando Etapa 1: Montagem e Carregamento dos Dados ---")
try:
    # Esta célula solicita autorização para acessar seus arquivos no Google Drive
    drive.mount('/content/drive')
    print("Google Drive montado com sucesso!")
except Exception as e:
    print(f"Drive já montado ou erro na montagem: {e}")

#
# --- AÇÃO NECESSÁRIA AQUI ---
#
# Substitua o caminho abaixo pelo caminho REAL para o seu arquivo de DADOS no Google Drive.
# Exemplo: '/content/drive/MyDrive/Dados/minha_tabela_landsat.csv'
#
data_path = '/content/drive/MyDrive/SEU_ARQUIVO_DE_DADOS.csv' # <-- MUDE ESTA LINHA

try:
    # O Pandas vai ler o arquivo de tabela (CSV, etc.) do caminho especificado
    df = pd.read_csv(data_path)
    print("\nArquivo de tabela lido com sucesso do Google Drive!")
    print("Abaixo, uma amostra dos seus dados para verificação:")
    print(df.head())
except FileNotFoundError:
    print(f"\n--- ERRO ---")
    print(f"O arquivo não foi encontrado no caminho: '{data_path}'.")
    print("Por favor, verifique se o caminho está 100% correto e se o arquivo existe no seu Drive.")
    # Interrompe a execução se o arquivo não for encontrado.
    raise
except Exception as e:
    print(f"\n--- ERRO ---")
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
    raise

# ===================================================================
# ETAPA 2: PREPARAR OS DADOS PARA A PINN
# (Conforme a estrutura da sua tabela: Bandas B1-B12 e 'landcover')
# ===================================================================
print("\n--- Iniciando Etapa 2: Processamento e Preparação dos Dados ---")

# 1. Identificar as colunas de bandas espectrais (as entradas do modelo)
band_columns = [f'B{i}' for i in range(1, 13)] # Colunas B1 a B12
r_observed = df[band_columns].values

# 2. Identificar os alvos (classes) a partir da coluna 'landcover'
target_names = df['landcover'].unique()
num_targets = len(target_names)
num_bands = len(band_columns)

# 3. Criar um mapa para converter nomes de classe (texto) em números (índices)
target_map = {name: i for i, name in enumerate(target_names)}
print(f"Alvos identificados ({num_targets} classes): {target_map}")

# 4. Criar as abundâncias verdadeiras (alpha_true)
# Assumimos que cada linha representa um pixel 100% puro da classe em 'landcover'
# Portanto, convertemos cada rótulo em um vetor "one-hot"
alpha_labels = df['landcover'].map(target_map).values
alpha_true = np.zeros((len(df), num_targets), dtype=np.float32)
alpha_true[np.arange(len(df)), alpha_labels] = 1.0

# 5. Converter tudo para tensores do PyTorch
r_observed_tensor = torch.tensor(r_observed, dtype=torch.float32)
alpha_true_tensor = torch.tensor(alpha_true, dtype=torch.float32)

print(f"\nDados prontos. Shape das entradas (r_observed_tensor): {r_observed_tensor.shape}")
print(f"Dados prontos. Shape das saídas (alpha_true_tensor): {alpha_true_tensor.shape}")

# 6. Extrair a matriz de endmembers (M) a partir dos dados puros
print("\nExtraindo assinaturas espectrais médias (endmembers) dos dados...")
M_true = np.zeros((num_bands, num_targets), dtype=np.float32)
for name, index in target_map.items():
    pure_pixels_mask = (df['landcover'] == name)
    mean_spectrum = df.loc[pure_pixels_mask, band_columns].mean().values
    M_true[:, index] = mean_spectrum

M_tensor = torch.tensor(M_true, dtype=torch.float32)
print("Matriz de endmembers extraída com sucesso.")

# ===================================================================
# ETAPA 3: DEFINIÇÃO DO MODELO PINN E TREINAMENTO
# ===================================================================
print("\n--- Iniciando Etapa 3: Definição e Treinamento do Modelo ---")

class EMIT_PINN(nn.Module):
    def __init__(self, num_bands, num_targets):
        super(EMIT_PINN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(num_bands, 128),
            nn.Tanh(),
            nn.Linear(128, 64),
            nn.Tanh(),
            nn.Linear(64, num_targets)
        )

    def forward(self, r_pixel):
        alpha_raw = self.net(r_pixel)
        # Força a restrição de não-negatividade
        alpha_non_negative = torch.relu(alpha_raw)
        # Força a restrição de soma-para-um
        alpha_sum_to_one = alpha_non_negative / (alpha_non_negative.sum(dim=1, keepdim=True) + 1e-8)
        return alpha_sum_to_one

# --- Hiperparâmetros de Treinamento ---
model = EMIT_PINN(num_bands, num_targets)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
lambda_physics = 0.5 # Peso para a perda da física (LMM)
lambda_data = 1.0    # Peso para a perda dos dados rotulados
epochs = 5000

# --- Loop de Treinamento ---
print(f"Iniciando o treinamento por {epochs} épocas...")
total_loss_history, data_loss_history, physics_loss_history = [], [], []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    # Predição de abundâncias para todo o dataset
    alpha_pred_data = model(r_observed_tensor)

    # 1. Perda de Dados (supervisionada)
    loss_data = nn.MSELoss()(alpha_pred_data, alpha_true_tensor)

    # 2. Perda de Física (auto-supervisionada, baseada no LMM)
    r_reconstructed = alpha_pred_data @ M_tensor.T
    loss_physics = nn.MSELoss()(r_reconstructed, r_observed_tensor)

    # 3. Perda Total Combinada
    total_loss = lambda_data * loss_data + lambda_physics * loss_physics

    # Armazenar histórico para plotagem
    total_loss_history.append(total_loss.item())
    data_loss_history.append(loss_data.item())
    physics_loss_history.append(loss_physics.item())

    # Backpropagation
    total_loss.backward()
    optimizer.step()

    if (epoch + 1) % 1000 == 0:
        print(f'Época [{epoch + 1}/{epochs}], Perda Total: {total_loss.item():.6f}, '
              f'Perda Dados: {loss_data.item():.6f}, Perda Física: {loss_physics.item():.6f}')

print("\nTreinamento concluído.")

# ===================================================================
# ETAPA 4: ANÁLISE DE PERFORMANCE E GERAÇÃO DE GRÁFICOS
# ===================================================================
print("\n--- Iniciando Etapa 4: Análise de Performance e Resultados ---")
model.eval() # Colocar modelo em modo de avaliação
with torch.no_grad():
    predicted_abundances = model(r_observed_tensor)

# Converter para NumPy para facilitar a análise e plotagem
alpha_true_np = alpha_true_tensor.numpy()
alpha_pred_np = predicted_abundances.numpy()

# --- Teste 1: Métrica Quantitativa de Erro ---
rmse = np.sqrt(mean_squared_error(alpha_true_np, alpha_pred_np))
print(f"\nRMSE (Raiz do Erro Quadrático Médio) geral: {rmse:.4f}\n")

# --- Gráfico 1: Curvas de Perda durante o Treinamento ---
plt.figure(figsize=(10, 6))
plt.plot(total_loss_history, label='Perda Total', color='black', linewidth=2)
plt.plot(data_loss_history, label=f'Perda de Dados (λ={lambda_data})', linestyle='--')
plt.plot(physics_loss_history, label=f'Perda da Física (λ={lambda_physics})', linestyle=':')
plt.xlabel('Época (Epoch)')
plt.ylabel('Valor da Perda (Loss)')
plt.title('Evolução das Funções de Perda Durante o Treinamento')
plt.legend()
plt.grid(True)
plt.yscale('log')
plt.show()

# --- Gráfico 2: Scatter Plot (Predito vs. Real) para cada Alvo ---
# Como os dados reais são apenas 0 ou 1, este gráfico mostrará a distribuição das predições para cada caso.
fig, axes = plt.subplots(3, 5, figsize=(20, 12))
fig.suptitle('Distribuição da Abundância Prevista vs. Real para Cada Alvo', fontsize=16)
for i, ax in enumerate(axes.flat):
    if i < num_targets:
        target_name = list(target_map.keys())[list(target_map.values()).index(i)]
        ax.scatter(alpha_true_np[:, i], alpha_pred_np[:, i], alpha=0.3)
        ax.set_title(f'Alvo: {target_name}')
        ax.set_xlabel('Abundância Real (0 ou 1)')
        ax.set_ylabel('Abundância Prevista')
        ax.grid(True)
        ax.set_xticks([0, 1])
        ax.set_ylim(0, 1)
    else:
        ax.set_visible(False)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Gráfico 3: Comparação de Abundâncias para um ÚNICO Pixel de Exemplo ---
pixel_idx_to_check = 0 # Analisando o primeiro pixel da tabela
labels = list(target_map.keys())
true_values = alpha_true_np[pixel_idx_to_check]
pred_values = alpha_pred_np[pixel_idx_to_check]

x = np.arange(len(labels))
width = 0.40

fig, ax = plt.subplots(figsize=(14, 7))
rects1 = ax.bar(x - width/2, true_values, width, label=f'Abundância Real (Pixel {pixel_idx_to_check} - Classe: {df["landcover"][pixel_idx_to_check]})')
rects2 = ax.bar(x + width/2, pred_values, width, label='Abundância Prevista pela PINN')

ax.set_ylabel('Fração de Abundância')
ax.set_title(f'Comparação da Composição do Pixel de Exemplo #{pixel_idx_to_check}')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.set_ylim(0, 1.1)
plt.tight_layout()
plt.show()

# --- Gráfico 4: Reconstrução da Assinatura Espectral para o mesmo Pixel ---
# Este gráfico mostra o quão bem a `loss_physics` funcionou.
original_spectrum = r_observed_tensor[pixel_idx_to_check].numpy()
reconstructed_spectrum = (predicted_abundances[pixel_idx_to_check] @ M_tensor.T).numpy()
band_wavelengths = [f'B{i}' for i in range(1, 13)] # Usando nomes das bandas no eixo X

plt.figure(figsize=(14, 7))
plt.plot(band_wavelengths, original_spectrum, label='Espectro Original Observado (r_observed)', color='black', marker='o')
plt.plot(band_wavelengths, reconstructed_spectrum, label='Espectro Reconstruído (M @ α_pred)', color='red', linestyle='--', marker='x')
plt.xlabel('Banda Espectral (Landsat)')
plt.ylabel('Reflectância (ou valor do pixel)')
plt.title(f'Qualidade da Reconstrução Espectral para o Pixel #{pixel_idx_to_check}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("\n--- Análise e geração de gráficos concluídas. ---")
```
````

````python
# -------------------------------------------------------------------
#                   *** BIBLIOTECAS NECESSÁRIAS ***
# -------------------------------------------------------------------
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.metrics import mean_squared_error

# ===================================================================
# ETAPA 1: MONTAR O GOOGLE DRIVE E CARREGAR SUA TABELA DE DADOS
# ===================================================================
print("--- Iniciando Etapa 1: Montagem e Carregamento dos Dados ---")
try:
    drive.mount('/content/drive')
    print("Google Drive montado com sucesso!")
except Exception as e:
    print(f"Drive já montado ou erro na montagem: {e}")

#
# --- AÇÃO NECESSÁRIA AQUI ---
#
# Substitua o caminho abaixo pelo caminho REAL para o seu arquivo de DADOS no Google Drive.
#
data_path = '/content/drive/MyDrive/GEE_Exports/assinaturas_espectrais_EMIT.csv' # <-- MUDE ESTA LINHA

try:
    df_long = pd.read_csv(data_path)
    print("\nArquivo de tabela (formato longo) lido com sucesso!")
    print("Amostra dos seus dados originais:")
    print(df_long.head())
except FileNotFoundError:
    print(f"\n--- ERRO ---")
    print(f"O arquivo não foi encontrado no caminho: '{data_path}'.")
    raise
except Exception as e:
    print(f"\n--- ERRO ---")
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
    raise

# ===================================================================
# ETAPA 2: PREPARAR OS DADOS PARA A PINN
# (Transformando dados do formato 'longo' para 'largo')
# ===================================================================
print("\n--- Iniciando Etapa 2: Transformação e Preparação dos Dados ---")

# 1. TRANSFORMAÇÃO (PIVOT)
# Esta é a etapa chave para o novo modelo de dados.
# Vamos transformar a tabela para que cada linha seja um pixel único.
# - 'index' define o que constitui uma linha única (um pixel e sua classe).
# - 'columns' define de onde vêm os novos nomes de coluna (as bandas).
# - 'values' define quais valores preencherão a tabela.
try:
    df_wide = df_long.pivot_table(
        index=['system:index', 'classe'],
        columns='banda',
        values='reflectancia'
    ).reset_index()

    # Fill NaN values with 0 after pivoting
    df_wide = df_wide.fillna(0)

    print("\nTabela transformada para o formato 'largo' com sucesso!")
    print("Amostra dos dados transformados:")
    print(df_wide.head())

    # Check for any remaining NaNs
    if df_wide.isnull().values.any():
        print("\n--- ATENÇÃO: Valores NaN ainda presentes após preenchimento. ---")
        print(df_wide.isnull().sum()[df_wide.isnull().sum() > 0])
        # Depending on the data, you might need a different strategy than fillna(0)
        # For now, we'll proceed but be aware of potential issues.

except Exception as e:
    print(f"\n--- ERRO AO TRANSFORMAR OS DADOS ---")
    print("Verifique se as colunas 'system:index', 'classe', 'banda', e 'reflectancia' existem no seu arquivo.")
    print(f"Erro original: {e}")
    raise

# 2. Extrair informações da tabela transformada
band_columns = [col for col in df_wide.columns if col not in ['system:index', 'classe']]
r_observed = df_wide[band_columns].values

target_names = df_wide['classe'].unique()
num_targets = len(target_names)
num_bands = len(band_columns)

target_map = {name: i for i, name in enumerate(target_names)}
print(f"\nAlvos identificados ({num_targets} classes): {target_map}")

# 3. Criar as abundâncias verdadeiras (alpha_true) a partir da coluna 'classe'
alpha_labels = df_wide['classe'].map(target_map).values
alpha_true = np.zeros((len(df_wide), num_targets), dtype=np.float32)
alpha_true[np.arange(len(df_wide)), alpha_labels] = 1.0

# 4. Converter tudo para tensores do PyTorch
r_observed_tensor = torch.tensor(r_observed, dtype=torch.float32)
alpha_true_tensor = torch.tensor(alpha_true, dtype=torch.float32)

print(f"\nDados prontos. Shape das entradas (r_observed_tensor): {r_observed_tensor.shape}")
print(f"Dados prontos. Shape das saídas (alpha_true_tensor): {alpha_true_tensor.shape}")

# 5. Extrair a matriz de endmembers (M)
print("\nExtraindo assinaturas espectrais médias (endmembers) dos dados...")
M_true = np.zeros((num_bands, num_targets), dtype=np.float32)
for name, index in target_map.items():
    pure_pixels_mask = (df_wide['classe'] == name)
    mean_spectrum = df_wide.loc[pure_pixels_mask, band_columns].mean().values
    M_true[:, index] = mean_spectrum

M_tensor = torch.tensor(M_true, dtype=torch.float32)
print("Matriz de endmembers extraída com sucesso.")


# ===================================================================
# ETAPA 3: DEFINIÇÃO DO MODELO PINN E TREINAMENTO
# (Nenhuma alteração necessária nesta seção)
# ===================================================================
print("\n--- Iniciando Etapa 3: Definição e Treinamento do Modelo ---")

class EMIT_PINN(nn.Module):
    def __init__(self, num_bands, num_targets):
        super(EMIT_PINN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(num_bands, 128), nn.Tanh(),
            nn.Linear(128, 64), nn.Tanh(),
            nn.Linear(64, num_targets)
        )

    def forward(self, r_pixel):
        alpha_raw = self.net(r_pixel)
        alpha_non_negative = torch.relu(alpha_raw)
        return alpha_non_negative / (alpha_non_negative.sum(dim=1, keepdim=True) + 1e-8)

model = EMIT_PINN(num_bands, num_targets)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
lambda_physics = 0.5
lambda_data = 1.0
epochs = 5000

print(f"Iniciando o treinamento por {epochs} épocas...")
total_loss_history, data_loss_history, physics_loss_history = [], [], []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    alpha_pred_data = model(r_observed_tensor)
    loss_data = nn.MSELoss()(alpha_pred_data, alpha_true_tensor)
    r_reconstructed = alpha_pred_data @ M_tensor.T
    loss_physics = nn.MSELoss()(r_reconstructed, r_observed_tensor)
    total_loss = lambda_data * loss_data + lambda_physics * loss_physics
    total_loss_history.append(total_loss.item())
    data_loss_history.append(loss_data.item())
    physics_loss_history.append(loss_physics.item())
    total_loss.backward()
    optimizer.step()
    if (epoch + 1) % 1000 == 0:
        print(f'Época [{epoch + 1}/{epochs}], Perda Total: {total_loss.item():.6f}')
print("\nTreinamento concluído.")

# ===================================================================
# ETAPA 4: ANÁLISE DE PERFORMANCE E GERAÇÃO DE GRÁFICOS
# (Nenhuma alteração necessária nesta seção)
# ===================================================================
print("\n--- Iniciando Etapa 4: Análise de Performance e Resultados ---")
model.eval()
with torch.no_grad():
    predicted_abundances = model(r_observed_tensor)

alpha_true_np = alpha_true_tensor.numpy()
alpha_pred_np = predicted_abundances.numpy()

rmse = np.sqrt(mean_squared_error(alpha_true_np, alpha_pred_np))
print(f"\nRMSE (Raiz do Erro Quadrático Médio) geral: {rmse:.4f}\n")

# --- Gráfico 1: Curvas de Perda ---
plt.figure(figsize=(10, 6))
plt.plot(total_loss_history, label='Perda Total', color='black', linewidth=2)
plt.plot(data_loss_history, label=f'Perda de Dados (λ={lambda_data})', linestyle='--')
plt.plot(physics_loss_history, label=f'Perda da Física (λ={lambda_physics})', linestyle=':')
plt.xlabel('Época (Epoch)'); plt.ylabel('Valor da Perda (Loss)'); plt.title('Evolução das Funções de Perda')
plt.legend(); plt.grid(True); plt.yscale('log'); plt.show()

# --- Gráfico 2: Scatter Plot por Alvo ---
num_rows = int(np.ceil(num_targets / 4))
fig, axes = plt.subplots(num_rows, 4, figsize=(20, 5 * num_rows))
fig.suptitle('Distribuição da Abundância Prevista vs. Real', fontsize=16)
axes = axes.flatten()
for i in range(num_targets):
    target_name = list(target_map.keys())[list(target_map.values()).index(i)]
    axes[i].scatter(alpha_true_np[:, i], alpha_pred_np[:, i], alpha=0.3)
    axes[i].set_title(f'Alvo: {target_name}')
    axes[i].set_xlabel('Abundância Real (0 ou 1)'); axes[i].set_ylabel('Abundância Prevista')
    axes[i].grid(True); axes[i].set_xticks([0, 1]); axes[i].set_ylim(0, 1)
for i in range(num_targets, len(axes)): axes[i].set_visible(False)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

# --- Gráfico 3: Comparação para um Pixel de Exemplo ---
pixel_idx_to_check = 0
pixel_info = df_wide.iloc[pixel_idx_to_check]
labels = list(target_map.keys())
true_values = alpha_true_np[pixel_idx_to_check]
pred_values = alpha_pred_np[pixel_idx_to_check]

x = np.arange(len(labels))
width = 0.40
fig, ax = plt.subplots(figsize=(14, 7))
ax.bar(x - width/2, true_values, width, label=f'Real (Pixel {pixel_info["system:index"]} - Classe: {pixel_info["classe"]})')
ax.bar(x + width/2, pred_values, width, label='Prevista pela PINN')
ax.set_ylabel('Fração de Abundância'); ax.set_title(f'Comparação da Composição do Pixel de Exemplo')
ax.set_xticks(x); ax.set_xticklabels(labels, rotation=45, ha="right")
ax.legend(); ax.grid(axis='y', linestyle='--', alpha=0.7); ax.set_ylim(0, 1.1); plt.tight_layout(); plt.show()

# --- Gráfico 4: Reconstrução Espectral do mesmo Pixel ---
original_spectrum = r_observed_tensor[pixel_idx_to_check].numpy()
reconstructed_spectrum = (predicted_abundances[pixel_idx_to_check] @ M_tensor.T).numpy()
band_wavelengths = band_columns

plt.figure(figsize=(14, 7))
plt.plot(band_wavelengths, original_spectrum, label='Espectro Original (r_observed)', color='black', marker='o')
plt.plot(band_wavelengths, reconstructed_spectrum, label='Espectro Reconstruído (M @ α_pred)', color='red', linestyle='--', marker='x')
plt.xlabel('Banda Espectral'); plt.ylabel('Reflectância'); plt.title(f'Qualidade da Reconstrução Espectral do Pixel de Exemplo')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.xticks(rotation=45); plt.show()

print("\n--- Análise e geração de gráficos concluídas. ---")
````
